# Build a Regression model in Keras

Importing the pandas and Numpy libraries

```python
import pandas as pd 
import numpy as np
```

Download the data and read it into a pandas dataframe

```python
concrete_data = pd.read_csv('concrete_data.csv')
concrete_data
```

```python
Check if the 
```

```python
concrete_data.isnull().sum()
```

Define X and Y for our dataset

```python
concrete_data_columns = concrete_data.columns
```

```python
X =(concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']])
```

```python
Y =(concrete_data['Strength'])
```

Train/Test dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split (X,Y,test_size =0.3,random_state =1)
```

Importing Keras libraries and packages

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

Define a function that defines our regression model

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation ='relu',input_dim=8))
    model.add(Dense(1))
    # compile model
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

Train the model

```python
model = regression_model()
model.fit(X_train,Y_train,epochs=50,verbose =1)
```

Evaluating the model

```python
Y_pred = model.predict(X_test)
Y_pred
```

Computing the mean squared error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,Y_pred)
```

Creating a list of 50 mean squared errors 

```python
mse_list = []
for n in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=1)
    model.fit(X_train, y_train, epochs=50, verbose=2)
    Y_pred = model.predict(X_test)
    sse = mean_squared_error(Y_test, Y_pred)
    mse_list.append(sse)
```

Mean and std of mse

```python
import statistics
mean = statistics.mean(mse_list)
print (mean)
```

```python
std = statistics.stdev(mse_list)
std
```
